```r
SPY_IEX_TEST <- riingo_iex_prices("SPY")
```

```r
SPY_IEX_TEST$m5_ret = SPY_IEX_TEST$close - SPY_IEX_TEST$open 
```

```r

SPY_water <- SPY_IEX_TEST %>%
  mutate(
    step = row_number(),
    cum_total = cumsum(m5_ret),
    direction = ifelse(m5_ret >= 0, "Up", "Down")
  )
```

```r
ggplot(SPY_water, aes(x = step)) +
  geom_col(aes(y = m5_ret, fill = direction), show.legend = FALSE) +
  geom_line(aes(y = cum_total), color = "black", linewidth = 1) +
  scale_fill_manual(values = c("Up" = "#2ecc71", "Down" = "#e74c3c")) +
  labs(
    title = "SPY Intraday Waterfall with Cumulative Return",
    x = "5-Minute Interval",
    y = "Return (Close - Open)"
  ) +
  theme_minimal(base_size = 14)
```
