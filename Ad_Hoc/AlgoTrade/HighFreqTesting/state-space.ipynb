```python
import pandas as pd
import tiingo
from tiingo import TiingoClient
import requests
import  json
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
config = {
    'session': True, # Recommended: Reuses HTTP session for better performance
    'api_key': os.getenv("TIINGO_API_KEY")
}

client = TiingoClient(config)
```

```python
df = client.get_dataframe(
        'SPY',
        frequency='1min',
        startDate='2025-06-01',
        endDate='2025-12-14'    
    )
```

```python
df.index = df.index.tz_convert('America/New_York')
```

```python
df['date'] = df.index
```

```python
print(f" Minimum date: {df['date'].min()} \n Maximum Date: {df['date'].max()} \n Count: {df['date'].count()}")
```
