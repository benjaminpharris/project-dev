```python
import torch

# Check if the MPS backend is available and built
if torch.backends.mps.is_available():
    print("✅ Success! PyTorch has access to the Metal (MPS) backend.")

    # Define the device to be the MPS GPU
    mps_device = torch.device("mps")

    # Create a tensor and move it to the GPU
    x = torch.ones(5, device=mps_device)
    print("\nHere is a tensor successfully created on the MPS device:")
    print(x)

    # Operations on the GPU will be much faster
    y = x * 2
    print("\nAnd here is the result of a computation on the device:")
    print(y)

else:
    print("❌ Failed. PyTorch cannot find the Metal (MPS) backend.")
    print("The installation may have defaulted to a CPU-only version.")
```
